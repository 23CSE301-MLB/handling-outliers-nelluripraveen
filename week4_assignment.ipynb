```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

iqr_outliers = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))

print("IQR Outlier counts:")
print(iqr_outliers.sum())
```

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df.select_dtypes(include=np.number)))
z_score_outliers = (z_scores > 3)

print("Z-Score Outlier counts:")
print(pd.Series(np.sum(z_score_outliers, axis=0), index=df.select_dtypes(include=np.number).columns))
```

```python
df_no_outliers = df[(z_scores < 3).all(axis=1)]
print("Shape after removing Z-score outliers:", df_no_outliers.shape)
```

```python
df_winsor = df.copy()
for col in df.select_dtypes(include=np.number).columns:
    lower = Q1[col] - 1.5 * IQR[col]
    upper = Q3[col] + 1.5 * IQR[col]
    df_winsor[col] = np.clip(df[col], lower, upper)

print("Winsorized Data Summary:")
print(df_winsor.describe())
```

```python
df_log = df.copy()
log_cols = [col for col in df.columns if (df[col] > 0).all() and df[col].dtype != 'object']

df_log[log_cols] = np.log1p(df_log[log_cols])

print("Log-Transformed Data Summary:")
print(df_log[log_cols].describe())
```
